```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dropout, GRU, Dense
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as ptch
%matplotlib inline
```

```python
##### Part I. Read in data and select features #####
TrainFrame = pd.read_csv('data/raw182_Training_Relabeled_Auto_25.csv')
TestFrame = pd.read_csv('data/raw91_Testing_Relabeled_Auto_25.csv')

inputColumns = [' ms_accelerometer_x',' ms_accelerometer_y',' ms_accelerometer_z']
outputColumns = ['outcome']

Xtrain = TrainFrame.loc[:, inputColumns]
Ytrain = TrainFrame.loc[:, outputColumns]

Xtest = TestFrame.loc[:, inputColumns]
Ytest = TestFrame.loc[:, outputColumns]
```

```python
Xtrain.head()
```

```python
batch_size = Xtrain.shape[0]
timesteps = Xtrain.shape[1]
data_dim = 1

model = Sequential()
model.add(GRU(input_shape=(timesteps, data_dim), units=50, return_sequences=True))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(loss="binary_crossentropy", optimizer="adam")
model.summary()
```

```python
Xtrain.shape
```

```python
model.fit(np.expand_dims(Xtrain, axis=2), Ytrain, epochs=5)
```

```python
y_pred = model.predict(np.expand_dims(Xtest, axis=2))
```

```python
from sklearn.metrics import mean_squared_error

print("MSE: {}".format(mean_squared_error(Ytest, y_pred)))
```
